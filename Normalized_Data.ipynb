```python
!pip install conllu

org_to_norm = {}

import conllu

def parse_conllu_file1(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.read()

    # Parse the CoNLL-U formatted data
    parsed_data = conllu.parse(data)

    # Extract tokens, lines, and folios
    tokens_info = []
    for sentence in parsed_data:
        for token in sentence:
            try:
                if 'misc' in token:
                    if "Norm" in token['misc']:
                        if "Orig" in token['misc']:
                            x = token['misc']['Orig']
                            if x not in org_to_norm:
                                org_to_norm[x] = ""


                        else:
                            x = token["form"]
                            if x not in org_to_norm:
                                org_to_norm[x] = ""
            except:
                continue
            
            
    return org_to_norm

def parse_conllu_file2(file_path):
    
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.read()

    # Parse the CoNLL-U formatted data
    parsed_data = conllu.parse(data)

    # Extract tokens, lines, and folios
    tokens_info = []
    for sentence in parsed_data:
        for token in sentence:
            try:
                if "misc" in token:
                    if "Norm" in token['misc']:
                        if "Orig" in token['misc']:
                            x = token['misc']['Orig']
                            y = token['misc']['Norm']
                            if y not in org_to_norm[x]:
                                org_to_norm[x]+=y+"#"
                                


                        else:
                            x = token["form"]
                            y = token['misc']['Norm']
                            if y not in org_to_norm[x]:
                                org_to_norm[x]+=y+"#"
                                
            except:
                continue

# Example usage
file_path = "book5.conllu"
parse_conllu_file1(file_path)
parse_conllu_file2(file_path)



for k in org_to_norm:
    org_to_norm[k] = org_to_norm[k][:-1]
    
    if "¶" in k:
        i = k.index("¶")
        t = []
        
        for c in org_to_norm[k]:
            t.append(c)
        
        hash_count = t[:i+1].count("#")
        
        t.insert(i+hash_count,"¶")
        org_to_norm[k] = "".join(t)

org_to_norm_final = {}

for k in org_to_norm:
    key = k
    
    if "¶·" in key:
        key = key.replace("¶·","¶")
    
    if "·" in key and "¶·" not in key:
        key = key.replace("·","#")
        
    
    if key[-1] == "¶" and len(key)>1:
        key = key[:-1]
        x = list(org_to_norm[k])
        x.remove("¶")
        value = "".join(x)
        org_to_norm_final[key] = value
    org_to_norm_final[key] = org_to_norm[k]


            
            
# Print the extracted information

```
